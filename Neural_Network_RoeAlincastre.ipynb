# **What is Neural Network (NN)**
A neural network (NN) is a computational model inspired by the way biological neural networks, like the human brain, process information. It is made up of layers of interconnected nodes (also called neurons), which are the building blocks of the network. These neurons simulate the behavior of brain neurons by receiving input, processing it, and passing the output to the next layer of neurons.

In a neural network:

1. Input layer receives the raw data.
2. Hidden layers (one or more) process the information by applying transformations, such as weighted sums and activation functions.
3. Output layer produces the final result, like a classification or regression output.

Neural networks learn from data by adjusting the weights (connections between neurons) and biases during training using an algorithm called backpropagation. This allows the network to minimize the error or loss in its predictions.

These models are widely used in various tasks such as image recognition, natural language processing, and even game playing, making them a fundamental tool in modern artificial intelligence and machine learning.

# **Understanding Neural Network:**

To understand NNs, students should be familiar with several key terms and concepts. These terms will help them grasp how neural networks function and how learning takes place within them. Here is a list of essential terms:

Neuron:

A neuron is the fundamental unit of a neural network. It mimics the function of biological neurons, where each node receives input, processes it, and produces output. In the neural network context:

•	Each neuron receives input values from the previous layer (or input layer), multiplies them by corresponding weights, sums them up, and adds a bias.

•	The neuron then applies an activation function to this sum to decide whether to "activate" and pass the signal to the next layer.

#	z **= W.X+b**

•	W is the weight vector,

•	X is the input vector,

•	b is the bias term.


Weights:

* Weights represent the strength of the connections between neurons. The network learns these weights during training.
* A high weight signifies that the corresponding input has a greater impact on the neuron’s output, while a low (or negative) weight reduces the impact.
* In a feedforward neural network, weights are learned in such a way that they minimize the error between the predicted output and the actual output.


Bias:

* The bias allows the model to fit the data better by shifting the activation function. It is like a constant added to the input of a neuron before applying the activation function.
* Without biases, the neuron's output might be restricted, especially in cases where all inputs are zero. Bias helps ensure that the model can approximate a wider range of functions.


Layers:

Neural networks are organized into different layers:
* Input Layer: Receives the raw data. For example, in an image recognition task, each pixel value of an image could be an input.

* Hidden Layers: Intermediate layers between the input and output layers. Each hidden layer transforms the input data and passes it to the next layer. These layers are where most of the learning and representation happens.
  * The more hidden layers and neurons in each layer, the deeper and potentially more powerful the model, though this can lead to more complex computations and a greater risk of overfitting.
* Output Layer: Produces the final predictions or classifications. In a classification task, the output layer would contain one neuron for each possible class, and each neuron’s output represents the probability that the input belongs to that class.


Activation Function:

The activation function introduces non-linearity into the model, which allows neural networks to approximate complex functions and solve problems that linear models cannot. Non-linear activation functions are crucial because without them, the entire neural network would behave like a linear function, no matter how many layers there are. There are three types of activation function shown below.

* Sigmoid: This function transforms z into a number between 0 and 1. It is particularly useful in the final layer of a neural network when dealing with binary classification problems, as it maps the output to a probability.

* ReLU(Rectified Linear Unit): It outputs the input directly if it is positive; otherwise, it outputs 0. ReLU is popular in deep learning because it helps networks learn faster and avoids vanishing gradient problems.
* Tanh (Hyperbolic tangent): Similar to sigmoid but outputs values between -1 and 1, providing stronger gradients than sigmoid for negative values

Feedforward:

* In a feedforward neural network, the data moves in one direction—from the input layer, through the hidden layers, and to the output layer.
* During this process, the network performs matrix multiplication and passes data through activation functions at each layer to produce the final output.
* There is no looping or backward connections in the feedforward process.


Loss Function:

The loss function measures how well the neural network’s predictions match the true data labels (in supervised learning).

* Mean Squared Error (MSE): Used for regression tasks where the goal is to minimize the difference between predicted and actual values.

* Cross-Entropy Loss: Used in classification problems to measure the difference between the predicted probability distribution and the true distribution. It’s highly effective for multi-class classification tasks.


image.png

Back Propagation:

Backpropagation is the key algorithm used for training neural networks. It calculates the gradient of the loss function with respect to each weight by applying the chain rule of calculus.
* The error from the output is propagated backward through the network to update the weights in a way that minimizes the loss.

The process involves:
1.	A forward pass to compute the output.
2.	A backward pass to compute the gradients of the loss concerning weights and biases.
3.	The update step, where weights are updated using gradient descent.

Gradient Descent:

Gradient descent is an optimization algorithm used to minimize the loss function by iteratively adjusting the weights in the direction of the steepest descent (negative gradient).
Variants of gradient descent include stochastic gradient descent (SGD), which updates the weights after each training example, and mini-batch gradient descent, which updates after processing a small subset of the data.


image.png

Learning Rate:

•	The learning rate 𝜂  is a hyperparameter that controls the size of the steps taken during gradient descent.
•	A high learning rate might cause the model to converge too quickly or oscillate around the minimum, while a low learning rate can lead to very slow convergence.


Epoch:

An epoch refers to one complete cycle where the entire training dataset is passed through the network once. Multiple epochs are usually required to train a model sufficiently, as weights are updated after each batch or example.

Batch Size:

* Batch size is the number of training examples processed before the model's weights are updated.
* Stochastic Gradient Descent uses a batch size of 1 (updating after each training example), while batch gradient descent uses the entire dataset to update the weights.


Overfitting:

* Overfitting occurs when the model performs well on training data but fails to generalize to unseen data. This typically happens when the network is too complex (too many layers or neurons) or is trained for too long.
* Symptoms of overfitting include low training loss but high validation loss.


Regularization:

Regularization techniques are used to prevent overfitting by reducing model complexity.
Common methods include:
* L2 Regularization (Weight Decay): Adds a penalty to large weights to prevent them from growing too large.
* Dropout: Randomly drops neurons during training, forcing the network to learn more robust features.
* Early Stopping: Stops the training process once the performance on the validation set stops improving.


Epoch vs. Iteration:

An epoch is a single pass through the entire training dataset, while an iteration is one step of the gradient update (which could involve a batch of data, depending on the batch size). If the dataset is divided into smaller batches, there are multiple iterations in each epoch.

# Part 2: Neural Network Implementation [50 marks]

The core of the lab is the neural network implementation. Some portions of the code are provided, while other sections are left as graded activities for the students to complete.

Activity 3 [10 points]: Define W1, b1, W2 and b2

```python
# Initialize weights and biases
def initialize_parameters(input_size, hidden_size, output_size):
    # Activity3: Define W1, b1, W2 and b2 [10 marks]
    # Your code:
    W1 = np.random.randn(input_size, hidden_size) * 0.01                        # Initialize weight matrix for input to hidden layer with small random values
    b1 = np.zeros((1, hidden_size))                                             # Initialize bias vector for hidden layer with zeros
    W2 = np.random.randn(hidden_size, output_size) * 0.01                       # Initialize weight matrix for hidden to output layer with small random values
    b2 = np.zeros((1, output_size))                                             # Initialize bias vector for output layer with zeros

    return W1, b1, W2, b2
```

Activity 4 [10 points]: Implement the ReLU activation function. You will be graded on the correct implementation of this functions.

Activity 5 [10 points]: Implement softmax. You will be graded on the correct implementation of this functions.

Instructions:

* Implement the ReLU activation function, which returns the input if it's positive, or 0 otherwise.
* Implement the softmax function, which normalizes the output of the network to probabilities.

```python
# Activation functions

# ReLU activation function
def relu(Z):
    # Activity 4: Implement the ReLU function here [5 marks]
    return np.maximum(0, Z)                                                     # Apply ReLU activation by setting negative values to zero

# Softmax activation function
def softmax(Z):
    # Activity 5: Implement the softmax function here [5 marks]
    exp_Z = np.exp(Z - np.max(Z, axis=1, keepdims=True))                        # Subtract max value for numerical stability
    return exp_Z / np.sum(exp_Z, axis=1, keepdims=True)                         # Normalize to get probability distribution
```

# Loss Function:

```python
# Cross-entropy loss function
def compute_loss(Y_true, Y_pred):
    m = Y_true.shape[0]  # Number of examples
    log_likelihood = -np.log(Y_pred[range(m), np.argmax(Y_true, axis=1)])
    return np.sum(log_likelihood) / m
```

Backward Propagation:

Activity 6-7 (20 points): Implement the backpropagation algorithm to compute the gradients for the weights and biases. You will be graded on the correctness of the gradient calculations for both the hidden and output layers.


Instructions:

* Compute the gradients for the weights and biases using the chain rule.
* Use the gradients to update the weights in the next step.

```python
# Backward Propagation

def backpropagation(X, Y, Z1, A1, A2, W2):
    m = X.shape[0]

    # Activity 6: Compute dZ2, dW2, db2 (gradients for the output layer) [10 marks]
    dZ2 = A2 - Y                                                                # Compute error at output layer
    dW2 = (1 / m) * np.dot(A1.T, dZ2)                                           # Compute gradient of weights for output layer
    db2 = (1 / m) * np.sum(dZ2, axis=0, keepdims=True)                          # Compute gradient of biases for output layer

    # Activity 7: Compute dA1, dZ1, dW1, db1 (gradients for the hidden layer) [10 marks]
    dA1 = np.dot(dZ2, W2.T)                                                     # Compute error for hidden layer
    dZ1 = dA1 * (Z1 > 0)                                                        # Compute gradient of activation function (ReLU derivative)
    dW1 = (1 / m) * np.dot(X.T, dZ1)                                            # Compute gradient of weights for hidden layer
    db1 = (1 / m) * np.sum(dZ1, axis=0, keepdims=True)                          # Compute gradient of biases for hidden layer

    return dW1, db1, dW2, db2
```

Activity 8 (10 points): Implement the parameter update function. Use gradient descent to update the weights and biases based on the gradients calculated during backpropagation.

Instructions:

* Use the gradients returned from backpropagation to update W1, b1, W2, and b2 with a learning rate.
* You will be graded on correct weight and bias updates using the provided learning rate.

```python
# Parameter Update

def update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate):
    # Activity 8: Implement weight and bias updates using gradient descent [10 marks]
    W1 -= learning_rate * dW1                                                   # Update weights for hidden layer
    b1 -= learning_rate * db1                                                   # Update biases for hidden layer
    W2 -= learning_rate * dW2                                                   # Update weights for output layer
    b2 -= learning_rate * db2                                                   # Update biases for output layer
    return W1, b1, W2, b2
```

# Part 3: Training the Neural Network [20 marks]

In this section, you will bring together the components implemented in Part 2 to train the neural network. The majority of this part is pre-written, except for a portion of the forward propagation loop, which you will need to fill in.

Activity 9 (10 points): Implement the forward propagation in the training loop. You need to ensure that the network is correctly passing the input through the layers and returning the output probabilities.

Activity 10 (10 points): use the forward_propagation and calculate loss

```python
#Forward Propagation

def forward_propagation(X, W1, b1, W2, b2):
    # Activity 09: Implement forward propagation logic here [10 marks]
    Z1 = np.dot(X, W1) + b1                                                     # Compute linear transformation for hidden layer
    A1 = relu(Z1)                                                               # Apply ReLU activation function
    Z2 = np.dot(A1, W2) + b2                                                    # Compute linear transformation for output layer
    A2 = softmax(Z2)                                                            # Apply softmax activation function to get class probabilities
    return Z1, A1, Z2, A2
```

```python
# Pre-written: Training function

def train(X, Y, input_size, hidden_size, output_size, epochs, learning_rate):
    W1, b1, W2, b2 = initialize_parameters(input_size, hidden_size, output_size)
    losses = []
    for epoch in range(epochs):

        #Activity 10: use the forward_propagation and calculate loss [10 marks]
        #Write code here:
        Z1, A1, Z2, A2 = forward_propagation(X, W1, b1, W2, b2)                 # Perform forward propagation
        loss = compute_loss(Y, A2)                                              # Compute loss
        losses.append(loss)                                                     # Store loss

        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss}")

        dW1, db1, dW2, db2 = backpropagation(X, Y, Z1, A1, A2, W2)
        W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate)

    return W1, b1, W2, b2,losses

# Train the model
input_size = X_train.shape[1]
hidden_size = 15
output_size = Y_train.shape[1]
learning_rate = 0.01
epochs = 10000

W1, b1, W2, b2, losses= train(X_train, Y_train, input_size, hidden_size, output_size, epochs, learning_rate)
```

# Part 4: Evaluation and Prediction and visualization [20 marks]

In this section, students will evaluate the model using a provided accuracy function. The evaluation is pre-written, so students can focus on the core learning objectives from the earlier sections.

```python
# Predict function
def predict(X, W1, b1, W2, b2):
    _, _, _, A2 = forward_propagation(X, W1, b1, W2, b2)
    predictions = np.argmax(A2, axis=1)
    return predictions

# Calculate accuracy
def calculate_accuracy(Y_true, Y_pred):

    true_labels = np.argmax(Y_true, axis=1)                                     # Activity 11: Complete the statement [5 marks] Extract true class labels from one-hot encoded Y
    accuracy = np.mean(true_labels == Y_pred)                                   # Activity 12: Complete the statement [5 marks] Compute mean of correctly predicted labels
    return accuracy

# Evaluate on test set
Y_pred_train = predict(X_train, W1, b1, W2, b2)
Y_pred_test = predict(X_test, W1, b1, W2, b2)

train_accuracy = calculate_accuracy(Y_train, Y_pred_train)
test_accuracy = calculate_accuracy(Y_test, Y_pred_test)

print(f"Training Accuracy: {train_accuracy * 100:.2f}%")
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
