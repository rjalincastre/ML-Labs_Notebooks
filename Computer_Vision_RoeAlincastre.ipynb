# **Introduction:**


**What is Computer Vision:**

Computer vision is a field of artificial intelligence (AI) that enables machines to interpret and understand the visual world. Using digital images from cameras, videos, and other sensors, computer vision algorithms can extract meaningful information and perform a variety of tasks like image classification, object detection, and image generation.

**Key Concepts in Computer Vision**

* Image Representation: Images are represented as grids (matrices) of pixels, where each pixel has a value that represents its intensity or color. For grayscale images, pixel values range between 0 (black) and 255 (white). For colored images, pixels are represented in three channels (RGB: Red, Green, Blue), each ranging from 0 to 255.

* Filters and Convolution: Convolution is a key operation in many computer vision algorithms. It involves applying a filter (also called a kernel) to an image to extract certain features, such as edges or textures. This is the basis for many modern computer vision techniques.

* Edge Detection: Edge detection involves identifying the boundaries of objects within an image. Techniques like Sobel, Prewitt, and Canny edge detection are commonly used to highlight areas where intensity changes sharply, which often corresponds to the edges of objects.

* Image Classification: In image classification, the goal is to assign an image to one of several predefined categories. Convolutional Neural Networks (CNNs) are the most popular deep learning architecture for this task. CNNs automatically learn important features from images like edges, textures, and shapes during training.

* Object Detection: Object detection goes beyond classification by not only recognizing objects but also determining their location in the image. Techniques like YOLO (You Only Look Once), SSD (Single Shot Multibox Detector), and Faster R-CNN are state-of-the-art methods for this.

* Segmentation: Segmentation is the process of partitioning an image into different regions or segments. Semantic segmentation classifies each pixel in the image into a predefined category, while instance segmentation distinguishes between different instances of the same object class.

* Face Recognition: Face recognition is a specific application of computer vision where algorithms detect and identify human faces in images or videos. It combines object detection and facial feature extraction to recognize and match faces.

* Optical Character Recognition (OCR): OCR is the process of extracting text from images. It is commonly used to digitize printed documents, enabling computers to read text from photographs, scanned documents, and even handwritten notes.

* Image Generation: Recent advances in AI have made it possible to generate realistic images from scratch using techniques like Generative Adversarial Networks (GANs). GANs consist of two neural networks (a generator and a discriminator) that work together to produce high-quality images that resemble real-world data.

**Applications of Computer Vision:**

* Autonomous Vehicles: Computer vision is crucial for self-driving cars to detect objects such as pedestrians, vehicles, traffic signals, and lane markings.
* Healthcare: In medical imaging, computer vision helps in diagnosing diseases by analyzing X-rays, MRIs, and CT scans.
* Security and Surveillance: Vision systems are used to monitor public spaces, detect unusual activity, and enhance security measures.
* Retail: Vision-based systems can track customer behavior, manage inventory, and improve the shopping experience through augmented reality (AR).

**Common Tools and Frameworks:**

* OpenCV: A popular open-source library for real-time computer vision tasks. It provides tools for image processing, object detection, and machine learning.
* TensorFlow and PyTorch: These are powerful frameworks for deep learning and are often used to build and train neural networks for vision-related tasks.
* Keras: A high-level neural networks API, running on top of TensorFlow, that simplifies the implementation of deep learning models, including those used for computer vision tasks.

**Challenges in Computer Vision**

* Occlusion: When parts of objects are hidden or obstructed, it can be difficult for the algorithm to identify the object correctly.
* Lighting Conditions: Variations in lighting can affect the quality of the images and the ability of models to make accurate predictions.
* Scale and Rotation Variability: Recognizing objects at different scales or from different angles can pose challenges for vision systems.
* Generalization: Models trained on specific datasets might not perform well in real-world settings where images differ in appearance and complexity.


Computer vision is a rapidly growing field with diverse applications across industries. With advances in AI, especially in deep learning, machines are increasingly capable of seeing and interpreting the world around them. As technology continues to evolve, we can expect computer vision to become an even more integral part of everyday life, improving areas such as automation, healthcare, and user interaction with technology.

**What are CNNs?**

Convolution is a mathematical operation that is essential in computer vision, especially for tasks like image processing and feature extraction. In deep learning, particularly in Convolutional Neural Networks (CNNs), convolution is used to automatically learn features such as edges, textures, patterns, and more from images. They are particularly effective in tasks like image classification, object detection, and segmentation.

**Key Components of CNNs:**

* Convolutional Layers: Apply filters to the input to extract features.
* Activation Functions: Introduce non-linearity (e.g., ReLU).
* Pooling Layers: Reduce spatial dimensions (e.g., Max Pooling).
* Fully Connected Layers: Perform high-level reasoning and classification.

**How Convolution Works:**

In the context of computer vision, convolution involves the following steps:

**Filter/Kernel:** A filter (also called a kernel) is a small matrix, usually 3x3 or 5x5, containing numerical values. The filter slides over the entire image and performs an element-wise multiplication of the filter and the corresponding portion of the image, followed by summing the results to produce a single value. This operation is repeated for each pixel in the image, generating a new matrix called a feature map or activation map.

Example of a 3x3 Filter (Edge Detection):



[-1 -1 -1]

[ 0  0  0]

[ 1  1  1]

This filter highlights horizontal edges in an image. When applied, areas with strong changes in pixel intensity (edges) will produce high values, while flat or smooth regions will produce low values.


**Strides and Padding:**

* Stride refers to how much the filter moves across the image. A stride of 1 means the filter moves one pixel at a time; a stride of 2 means it skips one pixel.

* Padding is adding extra borders to the image before applying convolution. This ensures that the filter can cover every part of the image, even the edges. Padding helps to control the output size of the feature map.
* Feature Extraction: The result of applying different filters to an image is a set of feature maps, each highlighting specific features. For example, one filter might detect horizontal edges, while another might detect vertical edges or textures. By stacking multiple convolutional layers in a CNN, deeper layers can capture more complex and abstract features (e.g., detecting eyes, wheels, etc.).

**Pooling:**

Reduces the Dimensionality of Feature Maps.
After applying convolution, the output feature maps are usually large, which increases the computational complexity. Pooling is used to reduce the dimensionality of these feature maps while retaining the important information.

**How Pooling Works:**

Pooling layers downsample the feature maps by summarizing the information in small regions of the map. This is done in a fixed-size window, such as 2x2 or 3x3, and results in smaller representations of the original feature maps.

* Max Pooling: This is the most common form of pooling. In a max pooling operation, the pooling window (e.g., 2x2) slides over the feature map, and the maximum value within the window is selected. This helps to retain the most prominent feature in each region while discarding less significant information.

* Average Pooling: In this method, instead of taking the maximum value, the average of all values in the pooling window is computed.

* Global Pooling: This is a more extreme version where the entire feature map is reduced to a single value, either using max or average pooling.

**Why Pooling is Important**

Pooling serves multiple purposes in CNNs:

* Dimensionality Reduction: Pooling reduces the spatial size of the feature maps, which decreases the number of parameters and computation in the network. This allows for faster training and inference.
* Invariance to Small Transformations: Max pooling, in particular, makes the model more robust to small translations and distortions in the image. This means that even if the object in the image is slightly shifted or rotated, the network can still recognize it.
* Prevents Overfitting: By summarizing features, pooling also helps reduce the chance of overfitting to the training data by generalizing the learned features.

**Fully Connected Layers: **

After the convolution and pooling layers have extracted the important features from the image, the final task is usually to classify the image into one or more categories. This is where fully connected layers come into play.

Fully connected (FC) layers work similarly to traditional neural networks where every neuron in one layer is connected to every neuron in the next layer. In a CNN, the feature maps produced by the convolution and pooling layers are flattened into a 1D vector, which is then fed into one or more fully connected layers.

* Flattening: After pooling, the 2D feature maps are converted into a 1D array (vector), which becomes the input to the fully connected layers. This vector represents the learned features from the earlier layers of the CNN.

          For example, if the feature maps have dimensions 7x7x64, flattening will produce a vector of size

                                 7 * 7 * 64 = 3136.

* Neurons and Weights: Each neuron in the fully connected layer has its own set of weights, which are multiplied by the input values and then passed through an activation function (such as ReLU or sigmoid) to introduce non-linearity.

* Final Output Layer: The final layer is often another fully connected layer where the number of neurons corresponds to the number of classes in the classification task. For example, in a digit classification task with 10 possible digits (0–9), the output layer will have 10 neurons. The output of this layer usually represents the probabilities of each class, which are calculated using a softmax function.

* Softmax Function: Softmax takes the raw output values from the final layer and converts them into probabilities. Each probability represents the likelihood that the input image belongs to a particular class, and all probabilities sum to 1.

**Why Fully Connected Layers Are Important**

* Classification: Fully connected layers are responsible for using the learned features to make predictions. The features extracted from the convolution and pooling layers are abstract representations of the image, and the fully connected layers help map these features to the output labels.
* Learning Non-linear Relationships: By applying non-linear activation functions between the layers, the fully connected layers can learn complex relationships between the input features and the output categories.

# **Step 2: Build a Simple CNN Model**

```python
import torch.nn as nn
import torch.nn.functional as F

# Define a simple CNN
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1)  # 3 input channels (RGB), 16 output channels, 3x3 kernel
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.pool = nn.MaxPool2d(2, 2)  # Max Pooling 2x2
        self.fc1 = nn.Linear(32 * 8 * 8, 128)  # Fully connected layer
        self.fc2 = nn.Linear(128, 10)  # 10 classes for CIFAR-10

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 32 * 8 * 8)  # Flatten the feature maps
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

net = SimpleCNN()
print(net)
```

# Activity 2: You are required to complete the code for the forward pass and try to modify the architecture (e.g., add more convolution layers or increase the kernel size).

```python
# For the above activity, write your code here with explanation

class ModifiedCNN(nn.Module):
    def __init__(self):
        super(ModifiedCNN, self).__init__()

        # First Convolutional Layer
        # Input: 3 channels (RGB), Output: 32 feature maps
        # Kernel size: 5x5, Padding: 2 to maintain image size
        self.conv1 = nn.Conv2d(3, 32, kernel_size=5, padding=2)
        self.bn1 = nn.BatchNorm2d(32)  # Normalize feature maps to stabilize training

        # Second Convolutional Layer
        # Input: 32 feature maps, Output: 64 feature maps
        # Kernel size: 5x5, Padding: 2
        self.conv2 = nn.Conv2d(32, 64, kernel_size=5, padding=2)
        self.bn2 = nn.BatchNorm2d(64)

        # Third Convolutional Layer (Added for deeper feature extraction)
        # Input: 64 feature maps, Output: 128 feature maps
        # Kernel size: 3x3, Padding: 1
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)

        # Max Pooling Layer
        # Reduces spatial dimensions by 2x2 (halves width & height)
        self.pool = nn.MaxPool2d(2, 2)

        # Fully Connected Layer 1
        # Input: Flattened feature maps (128 channels × 4 × 4)
        # Output: 256 neurons
        self.fc1 = nn.Linear(128 * 4 * 4, 256)

        # Dropout layer (prevents overfitting by randomly deactivating neurons)
        self.dropout = nn.Dropout(0.3)

        # Fully Connected Layer 2 (Final Output Layer)
        # Input: 256 neurons, Output: 10 (CIFAR-10 has 10 classes)
        self.fc2 = nn.Linear(256, 10)

    def forward(self, x):
        # Pass input through first convolutional layer + ReLU activation + BatchNorm + MaxPooling
        x = self.pool(F.relu(self.bn1(self.conv1(x))))

        # Second convolutional layer + ReLU activation + BatchNorm + MaxPooling
        x = self.pool(F.relu(self.bn2(self.conv2(x))))

        # Third convolutional layer + ReLU activation + BatchNorm + MaxPooling
        x = self.pool(F.relu(self.bn3(self.conv3(x))))

        # Flatten the feature maps from (batch_size, 128, 4, 4) → (batch_size, 128 * 4 * 4)
        x = x.view(-1, 128 * 4 * 4)

        # Pass through first fully connected layer + ReLU activation
        x = F.relu(self.fc1(x))

        # Apply dropout to prevent overfitting
        x = self.dropout(x)

        # Pass through the final fully connected layer (logits for 10 classes)
        x = self.fc2(x)

        return x

# Instantiate the model and print its architecture
net = ModifiedCNN()
print(net)
```

# **Step 3: Train the CNN Model (20 minutes)**

```python
import torch.optim as optim

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters(), lr=0.01)

# Training loop
for epoch in range(5):  # loop over the dataset multiple times
    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        inputs, labels = data

        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = net(inputs)
        loss = criterion(outputs, labels)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        if i % 200 == 199:    # Print every 200 mini-batches
            print(f'Epoch {epoch + 1}, Batch {i + 1}: Loss {running_loss / 200}')
            running_loss = 0.0

print('Finished Training')
```

**What are RNNs?**

Recurrent Neural Networks (RNNs) are designed to handle sequential data by maintaining a hidden state that captures information from previous inputs. They are widely used in tasks like language modeling, machine translation, and time series prediction.

**Long Short-Term Memory (LSTM) Networks**

LSTMs are a type of RNN that address the vanishing gradient problem, allowing them to capture long-term dependencies in data. They achieve this through gating mechanisms: input gate, forget gate, and output gate.

# Dataset: Character-Level Text Generation

# **Step 1: Import libraries.**

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
```

# **Step 2: Prepare Dataset**

```python
# Sample text data
text = "hello world this is a simple example of character level text generation using LSTM networks"

# Create a set of unique characters
chars = sorted(list(set(text)))
vocab_size = len(chars)

# Create mappings from characters to indices and vice versa
char_to_idx = {ch: idx for idx, ch in enumerate(chars)}
idx_to_char = {idx: ch for idx, ch in enumerate(chars)}

# Convert the text into a sequence of indices
input_seq = [char_to_idx[ch] for ch in text]
```

# **Step 3: Define LSTM Model**

```python
class CharLSTM(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, n_layers=2):
        super(CharLSTM, self).__init__()
        self.hidden_size = hidden_size
        self.n_layers = n_layers

        # LSTM layer
        self.lstm = nn.LSTM(input_size, hidden_size, n_layers, batch_first=True)
        # Fully connected layer
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x, hidden):
        # Pass through LSTM layer
        out, hidden = self.lstm(x, hidden)
        # Reshape output for the fully connected layer
        out = out.contiguous().view(-1, self.hidden_size)
        # Pass through fully connected layer
        out = self.fc(out)
        return out, hidden

    def init_hidden(self, batch_size):
        # Initialize hidden state and cell state with zeros
        weight = next(self.parameters()).data
        hidden = (weight.new(self.n_layers, batch_size, self.hidden_size).zero_(),
                  weight.new(self.n_layers, batch_size, self.hidden_size).zero_())
        return hidden
```

# **Step 4: Hyperparameter**

```python
# Hyperparameters
input_size = vocab_size  # One-hot encoding
hidden_size = 128
output_size = vocab_size
n_layers = 2
lr = 0.003

# Instantiate the model
model = CharLSTM(input_size, hidden_size, output_size, n_layers)

# Define the loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=lr)
```

# **Step 5: Model Training:**

```python
# One-hot encoding function
def one_hot_encode(sequence, n_labels):
    encoding = torch.zeros(len(sequence), n_labels)
    for i, idx in enumerate(sequence):
        encoding[i][idx] = 1
    return encoding

# Convert the entire input sequence to one-hot encoding
input_encoded = one_hot_encode(input_seq, vocab_size)

# Define sequence length
seq_length = 10  # Number of characters per sequence

# Create input and target sequences
def create_sequences(input_data, seq_length):
    inputs = []
    targets = []
    for i in range(len(input_data) - seq_length):
        inputs.append(input_data[i:i+seq_length])
        targets.append(input_data[i+seq_length])
    return inputs, targets

inputs, targets = create_sequences(input_encoded, seq_length)
```

# **Step 6: Train LSTM**

```python
# Number of epochs
epochs = 50

for epoch in range(1, epochs + 1):
    hidden = model.init_hidden(1)
    model.zero_grad()

    loss = 0
    # Iterate through the sequence
    for i in range(seq_length):
        input_char = inputs[0][i].unsqueeze(0).unsqueeze(0)  # Shape: (1, 1, vocab_size)
        output, hidden = model(input_char, hidden)
        # Get the index of the maximum value from the one-hot encoded target
        target_index = torch.argmax(targets[0][i]).long()
        # Calculate loss using the target index
        loss += criterion(output, target_index.unsqueeze(0))

    # Backpropagation
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item()/seq_length:.4f}')
```

# **Step 7: Generate Text with Model**

```python
def generate_text(model, start_str, generate_length=100):
    model.eval()
    hidden = model.init_hidden(1)
    input_chars = [char_to_idx[ch] for ch in start_str]
    input_encoded = one_hot_encode(input_chars, vocab_size).unsqueeze(0)  # Shape: (1, len, vocab_size)

    generated_str = start_str
    input_seq = input_encoded[:, -1, :].unsqueeze(0)  # Last character, shape: (1, vocab_size) # Change here

    for _ in range(generate_length):
        output, hidden = model(input_seq, hidden)
        _, top_idx = output.topk(1)
        char_idx = top_idx.item()
        generated_str += idx_to_char[char_idx]
        # Prepare the next input
        input_seq = one_hot_encode([char_idx], vocab_size).unsqueeze(0) # Change here


    print(f'Generated Text:\n{generated_str}')
```

# **Part 3: Implementing a HyBrid Model**

In this section, we will combine a Convolutional Neural Network (CNN) and a Recurrent Neural Network (RNN) to process a real dataset.

This hybrid model can be useful for tasks where you need to extract spatial features (like from images) and capture temporal or sequential information (like from time-series or sequential data).

For this example, we will use the MNIST handwritten digit dataset (a widely used dataset for image classification) and combine CNN to extract features from the images and RNN to capture any temporal dependencies that may exist in sequential data, such as image row-wise patterns. While images do not have a strict time-series component, we can simulate this by treating each row of pixels as a sequence of data.

Let’s dive into the example using PyTorch:

# Step 1: Import Libraries

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
```

# Step 2: Preparing the MNIST Dataset

# We’ll use the MNIST dataset, where each image is 28x28 pixels. We’ll treat each row of 28 pixels as a time step, so each image can be viewed as a sequence of 28 time steps, each with 28 features (pixels).

```python
# Define transformations for the dataset (normalizing the data)
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Load the MNIST dataset
train_dataset = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
test_dataset = datasets.MNIST(root='./data', train=False, transform=transform)

train_loader = DataLoader(dataset=train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=64, shuffle=False)
```

# Step 3: Defining the CNN-RNN Hybrid Model
# Here, we define a model that uses CNN layers to extract features from the image and an RNN (specifically, an LSTM) to capture sequential dependencies from the row-wise data.

```python
class CNN_RNN_Model(nn.Module):
    def __init__(self, num_classes=10):
        super(CNN_RNN_Model, self).__init__()

        # CNN Part
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2, padding=0)

        # RNN Part (LSTM)
        self.lstm = nn.LSTM(input_size=32*7, hidden_size=128, num_layers=2, batch_first=True)

        # Fully Connected Layer for classification
        self.fc = nn.Linear(in_features=128, out_features=num_classes)

    def forward(self, x):
        # CNN layers
        x = self.pool(torch.relu(self.conv1(x)))  # Output: [batch_size, 16, 14, 14]
        x = self.pool(torch.relu(self.conv2(x)))  # Output: [batch_size, 32, 7, 7]

        # Prepare data for LSTM
        x = x.permute(0, 2, 1, 3)  # Change to [batch_size, 7, 32, 7]
        x = x.reshape(x.size(0), x.size(1), -1)  # Flatten: [batch_size, 7, 32*7]

        # RNN layers
        lstm_out, (hn, cn) = self.lstm(x)  # LSTM output
        lstm_out = lstm_out[:, -1, :]  # Take output from last time step

        # Fully connected layer
        out = self.fc(lstm_out)

        return out
```

# Step 4: Training the Model
# We will define the training loop, where we use cross-entropy loss and stochastic gradient descent (SGD) as the optimizer.

```python
# Set device to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Instantiate model, loss function, and optimizer
model = CNN_RNN_Model(num_classes=10).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training function
def train_model(model, train_loader, criterion, optimizer, num_epochs=5):
    model.train()
    for epoch in range(num_epochs):
        running_loss = 0.0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            # Forward pass
            outputs = model(images)
            loss = criterion(outputs, labels)

            # Backward pass and optimization
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}')

    print("Training complete.")

# Train the model
train_model(model, train_loader, criterion, optimizer, num_epochs=5)
```

# Step 5: Testing the Model
# We will evaluate the model on the test dataset to check its accuracy.

```python
# Test the model
def test_model(model, test_loader):
    model.eval()  # Set to evaluation mode
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    print(f'Accuracy of the model on the test images: {100 * correct / total:.2f}%')

# Test the model
test_model(model, test_loader)
```

## **9. What is the shape of a single MNIST image?**  
A single MNIST image has a shape of (1, 28, 28). The 1 represents the number of color channels, as the images are grayscale. The 28x28 corresponds to the height and width of the image in pixels.

## **10. How does normalization affect the pixel values in the images?**  
Normalization rescales the pixel values from their original range of 0 to 255 to a standardized range, typically between -1 and 1 (using a mean and standard deviation of 0.5). This process helps stabilize the training process by ensuring consistent input distributions, reducing the impact of extreme values, and improving the convergence speed of the model.

## **11. How does the output size change after each CNN and pooling operation?**  
The initial image size is (1, 28, 28). After passing through the first convolutional layer (`conv1`), the output becomes (16, 28, 28), since padding maintains the spatial dimensions. A max-pooling operation reduces this to (16, 14, 14). The second convolutional layer (`conv2`) then expands it to (32, 14, 14), followed by another pooling operation that reduces it further to (32, 7, 7).

## **12. Why is it necessary to reshape the output of the CNN before feeding it into the LSTM?**  
The LSTM expects input in the format (batch_size, sequence_length, input_size). However, the output from the CNN has the shape (batch_size, 32, 7, 7), which is not suitable for sequential processing. To convert it into a sequential form, the output is reshaped to (batch_size, 7, 32*7). This transformation treats each row of the feature map as a time step, allowing the LSTM to process the spatial features in a sequential manner.
